<b>Birthday Paradox </b>

How many people are required at a party so 2 people have the birthday in the same day with a probability higher than 50%

```python
import seaborn as sbs; sbs.set()
import matplotlib.pyplot as plt
from matplotlib import mlab
%matplotlib inline
```

```python
def get_probability(k):
    favorable_events = 365
    for i in range(1, k):
        favorable_events *= (365 - i)
    
    return 1 - favorable_events / (365 ** k)
```

```python
fig = plt.figure(figsize=(20,10))
days = [i for i in range(1, 50)]
solutions = [get_probability(nr_day) for nr_day in days]
plt.plot(solutions, days)
plt.axis([0, 1, 1, 50])
plt.title('Probability based on number of people')
plt.xlabel('Probability')
plt.ylabel('Number of people')
plt.show()
```

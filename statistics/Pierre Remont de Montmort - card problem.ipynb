Pierre Remond de Montmort was a French mathematician who studied probability in the context of gambling and wrote a treatise devoted to the analysis of various card games. He posed the following problem in 1708, based on a card game called Treize.

Consider a well-shuffled deck of n cards, labeled 1 through n. You flip over the cards one by one, saying the numbers 1 through n as you do so. You win the game if, at some point, the number you say aloud is the same as the number on the card being flipped over (for example, if the 7th card in the deck has the label 7). What is the probability of winning?

```python
import numpy as np
from numpy import random
```

```python
import matplotlib.pyplot as plt
from matplotlib import mlab
%matplotlib inline
```

Generate a permutation and check how probability changes when increasing number of cards

```python
def get_win_probability(number_of_cards):
    
    sorted_values = [i for i in range(1, number_of_cards)]
    number_of_cards_at_correct_index = 0
    for i in range(1000):
        permutation = random.permutation(sorted_values)
        number_of_cards_at_correct_index += 1 if np.sum(permutation == sorted_values) != 0 else 0

    return number_of_cards_at_correct_index/1000
```

```python
fig = plt.figure(figsize=(20,10))
plt.title("Probability to win the game", fontname='Monospace', fontdict={'fontsize':14}, y=1.03)
cards_count = [nr_cards for nr_cards in range(20, 1000, 25)]

plt.scatter(cards_count, [get_win_probability(nr_cards) for nr_cards in cards_count])
plt.ylabel("Probability", fontsize = 14)
plt.xlabel("Cards", fontsize = 14)
```
